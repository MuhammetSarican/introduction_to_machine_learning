```python
import random

kazandiniz=0
kaybettiniz=0
tur=0
keci=0
kapi=[1,2,3]

dongu=int(input("Döngü sayısını giriniz:\n"))
secim=input("Kapınızı değiştirmek istiyor musunuz?[y-n]\n")

while tur<dongu:
    monty_hall=[1,1,1]

    a=random.randint(0,2)
    monty_hall[a]=0

#     tut_kapi=int(input("Tutulan kapıyı girin:(1-3)\n"))
    tut_kapi=int(random.randint(1,3))
    #print("\nSeçilen Kapı: ",tut_kapi)

    for index, value in enumerate(monty_hall):
        if value==1 and index!=tut_kapi-1:
            #print(index+1," Numaralı Kapı:'Keçi'")
            keci=index+1
            break
            

    if secim=="y" or secim=="Y":
        for a in kapi:
            if a!=tut_kapi and a!=keci:
                tut_kapi=a
                #print("\nDeğiştirilen Kapı: ",tut_kapi)
                break

    if monty_hall[tut_kapi-1]==0:
        #print("\n'Kazandınız'")
        kazandiniz+=1
    else:
        #print("\n'Kaybettiniz'")
        kaybettiniz+=1
    tur+=1

    

    #print(monty_hall[0],":",monty_hall[1],":",monty_hall[2])

print ("\nKazanılan: ",kazandiniz," Kaybedilen: ",kaybettiniz,"Tur:",tur,"\n\nKazanma Olasılığı:\n%",(kazandiniz/tur)*100)
```
