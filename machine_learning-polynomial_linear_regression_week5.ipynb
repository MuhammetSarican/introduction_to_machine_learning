```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

data=pd.read_csv("polynomial-regression.csv",sep = ";")

x=data.car_max_speed.values.reshape(-1,1)
y=data.car_price.values.reshape(-1,1)

plt.scatter(x,y)
plt.xlabel("car_max_speed")
plt.ylabel("car_price")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(x,y)

y_head=lr.predict(x)

plt.scatter(x,y)
plt.plot(x,y_head,color="red",label ="linear")
```

```python
from sklearn.preprocessing import PolynomialFeatures
p_r=PolynomialFeatures(degree=2)

x_pol=p_r.fit_transform(x)

linear_reg2=LinearRegression()
linear_reg2.fit(x_pol,y)

y_head2=linear_reg2.predict(x_pol)

plt.scatter(x,y)
plt.plot(x,y_head,color="red",label ="linear")
plt.plot(x,y_head2,color= "green",label = "poly")
```

```python
from sklearn.metrics import r2_score
print("r_score for linear regression",r2_score(y,y_head))
print("r_score for polynomial regression",r2_score(y,y_head2))
```
